```python
# python: xa['A'] + xa['B']
# Subscript operators call __getitem__
#frame.py
    def __getitem__(self, key):
        if type(key) is XArray:
            return self._row_selector(key)
        elif type(key) is list:
            return self.select_columns(key)
        elif type(key) is str:
            return self.select_column(key)
        elif type(key) is int:
            if key < 0:
                key = len(self) + key
            if key >= len(self):
                raise IndexError("XFrame index out of range")
            return list(XFrame(_impl = self.__impl__.copy_range(key, 1, key+1)))[0]
        elif type(key) is slice:
            start = key.start
            stop = key.stop
            step = key.step
            if start is None:
                start = 0
            if stop is None:
                stop = len(self)
            if step is None:
                step = 1
            # handle negative indices                                                                                                  
            if start < 0:
                start = len(self) + start
            if stop < 0:
                stop = len(self) + stop
            return XFrame(_impl = self.__impl__.copy_range(start, step, stop))
        else:
            raise TypeError("Invalid index type: must be XArray, int, list, slice, or str")

```

```python
# stdXArrayImpl.py
def vector_operator(self, other, op):                                                                                              
        """                                                                                                                            
        Performs an element-wise operation on the two RDDs.                                                                            
        """                                                                                                                            
        self._entry(other, op)                                                                                                         
        res_type = self.elem_type                                                                                                      
        pairs = safeZip(self.rdd, other.rdd)                                                                                           
        if op == '+':                                                                                                                  
            res = pairs.map(lambda x: x[0] + x[1], preservesPartitioning=True)                                                         
        elif op == '-':                                                                                                                
            res = pairs.map(lambda x: x[0] - x[1], preservesPartitioning=True)                                                         
        elif op == '*':                                                                                                                
            res = pairs.map(lambda x: x[0] * x[1], preservesPartitioning=True)                                                         
        elif op == '/':                                                                                                                
            res = pairs.map(lambda x: x[0] / x[1], preservesPartitioning=True)                                                         
        elif op == '<':                                                                                                                
            res = pairs.map(lambda x: x[0] < x[1], preservesPartitioning=True)                                                         
            res_type = int
        elif op == '>':
            res = pairs.map(lambda x: x[0] > x[1], preservesPartitioning=True)
            res_type = int
        elif op == '<=':
            res = pairs.map(lambda x: x[0] <= x[1], preservesPartitioning=True)
            res_type = int
        elif op == '>=':
            res = pairs.map(lambda x: x[0] >= x[1], preservesPartitioning=True)
            res_type = int
        elif op == '==':
            res = pairs.map(lambda x: x[0] == x[1], preservesPartitioning=True)
            res_type = int
        elif op == '!=':
            res = pairs.map(lambda x: x[0] != x[1], preservesPartitioning=True)
            res_type = int
        elif op == '&':
            res = pairs.map(lambda x: x[0] and x[1], preservesPartitioning=True)
            res_type = int
        elif op == '|':
            res = pairs.map(lambda x: x[0] or x[1], preservesPartitioning=True)
            res_type = int
        else:
            raise NotImplementedError(op)
        self._exit()
        return self._rv(res, res_type)

 
```

### Illustration of Stack Implementation

```python
# Top Level
def stack(self, column_name, new_column_name=None, drop_na=False):
        # validate column_name                                                                                                         
        column_name = str(column_name)
        if column_name not in self.column_names():
            raise ValueError("Cannot find column '" + str(column_name) + "' in the XFrame.")

        stack_column_type =  self[column_name].dtype()
        if (stack_column_type not in [dict, array.array, list]):
            raise TypeError("Stack is only supported for column of dict/list/array type.")

        if (new_column_name != None):
            if stack_column_type == dict:
                if (type(new_column_name) is not list):
                    raise TypeError("new_column_name has to be a list to stack dict type")
                elif (len(new_column_name) != 2):
                    raise TypeError("new_column_name must have length of two")
            else:
                if (type(new_column_name) != str):
                    raise TypeError("new_column_name has to be a str")
                new_column_name = [new_column_name]

            # check if the new column name conflicts with existing ones                                                                
            for name in new_column_name:
                if (name in self.column_names()) and (name != column_name):
                    raise ValueError("Column with name '" + name + "' already exists, pick a new column name")
        else:
            if stack_column_type == dict:
                new_column_name = ["",""]
            else:
                new_column_name = [""]

        # infer column types                                                                                                           
        # TODO do this with head_as_list                                                                                               
        head_row = XArray(self[column_name].head(100)).dropna()
        if (len(head_row) == 0):
            raise ValueError("Cannot infer column type because there is not enough rows to infer value")
        if stack_column_type == dict:
            # infer key/value type                                                                                                     
            keys = []; values = []
            for row in head_row:
                for val in row:
                    keys.append(val)
                    if val != None: values.append(row[val])

            new_column_type = [
                infer_type_of_list(keys),
                infer_type_of_list(values)
            ]
        else:
            values = [v for v in itertools.chain.from_iterable(head_row)]
            new_column_type = [infer_type_of_list(values)]

        with impl_context():
            if stack_column_type == dict:
                return XFrame(_impl=self.__impl__.stack_dict(column_name, new_column_name, new_column_type, drop_na))
            else:
                return XFrame(_impl=self.__impl__.stack_list(column_name, new_column_name, new_column_type, drop_na))
```

```python
# Implementation Level
def stack_dict(self, column_name, new_column_names, new_column_types, drop_na):
    self._entry(column_name, new_column_names, new_column_types, drop_na)
    col_num = self.col_names.index(column_name)
    def subs_row(row, col, key, val):
        new_row = list(row)
        new_row[col] = key
        new_row.insert(col+1, val)
        return new_row
    def stack_row(row, col, drop_na):
        res = []
        for key, val in row[col].iteritems():
            if drop_na and is_missing_or_empty(val): continue
            res.append(subs_row(row, col, key, val))
        if len(res) > 0 or drop_na:
            return res
        return [subs_row(row, col, None, None)]
    res = self.rdd.flatMap(lambda row: stack_row(row, col_num, drop_na))

    column_names = list(self.col_names)
    new_name_k = new_column_names[0]
    if new_name_k == '':
        new_name_k = name_col(column_names, 'K')
    column_names[col_num] = new_name_k
    new_name_v = new_column_names[1]
    if new_name_v == '':
        new_name_v = name_col(column_names, 'V')
    column_names.insert(col_num + 1, new_name_v)
    column_types = list(self.column_types)
    column_types[col_num] = new_column_types[0]
    column_types.insert(col_num + 1, new_column_types[1])
    self._exit(res, column_names, column_types)
    return self._rv(res, column_names, column_types)
```

### Modified 3-Jaccard

```python
# Computes the jaccard similarity based on a set of claims.                                                                            
# Relationships between providers are formed thorugh shared patients.                                                                  
# The provider-patient relationship can be weighted or unweighted, but the data need                                                   
#   not be unique -- repeated linkages are summed.                                                                                     
class Modified3Jaccard:
    """ Create a score for each pair of providers that measures how similar they are,                                                  
        based on the number of patients they share.                                                                                    
                                                                                                                                       
        The similarity can also be weighted by the provider score, so that a high                                                      
        scoring provider will         share this score with the providers with which                                                   
        they share patients. """

    # If weight_col is not given, then weights are assumed to be 1                                                                     
    # Either way, there may be one or more provider/patient entries, whose                                                             
    #    weights are summed up                                                                                                         
    def __init__(self, claims, provider_col, patient_col, weight_col=None):
        self.claims = SFrame(claims)
        self.claims.rename({provider_col: 'provider',
                            patient_col: 'patient'})
        self.provider_col = provider_col
        self.patient_col = patient_col
        if weight_col is None:
            weight_col = 'weight'
            self.claims[weight_col] = 1.0
        self.weight_col = weight_col


    def make_similarity(self, k=10):
        def divide(num, denom):
            return num / denom if num != 0 else 0.0

        def regularize(num, denom, mean, k):
            return (num + k * mean) / (denom + k)

        if len(self.claims) == 0:
            # no claims at all                                                                                                         
            return None

        claim_counts = self.claims.groupby(['provider', 'patient'],
                                           {'claim-count': SUM(self.weight_col)})
        # provider patient claim-count                                                                                                 
        # these are providers and the patients they have seen                                                                          
        #print 'claim_counts'                                                                                                          
        #print claim_counts                                                                                                            

        # make two copies of the claim counts so that we can join                                                                      
        provider_patient = SFrame(claim_counts)
        provider_patient.rename({'provider': 'provider1', 'claim-count': 'claim-count1'})
        # provider1 claim-count1 patient                                                                                               
        #print 'provider_patient'                                                                                                      
        #print provider_patient                                                                                                        

        patient_provider = SFrame(claim_counts)
        patient_provider.rename({'provider': 'provider2', 'claim-count': 'claim-count2'})
        # provider2 patient claim-count2                                                                                               

        prov_prov = patient_provider.join(provider_patient, on='patient')
        # provider1 count1 provider2 count2 patient                                                                                    

        # discard redundant provider/provider pairs                                                                                    
        prov_prov = prov_prov[prov_prov['provider1'] > prov_prov['provider2']]

        # give up if there is nothing left                                                                                             
        if len(prov_prov) == 0:
            # no claims with common patients                                                                                           
            return None

        # do the min for all provider-provider pairs                                                                                   
        prov_prov['numerator-term'] = prov_prov.apply(
            lambda row: 2.0 * min(row['claim-count1'], row['claim-count2']))
        del prov_prov['claim-count1']
        del prov_prov['claim-count2']
        # provider1 provider2 patient numerator-term                                                                                   

        prov_prov = prov_prov.join(patient_claim_count, on='patient')
        prov_prov.rename({'claim-count': 'denominator-term'})
        # provider1 provider2 patient numerator-term denominator-term                                                                  

        # sum numerator-term and denominator-term over patient for each provider pair separately                                       
        similarity = prov_prov.groupby(['provider1', 'provider2'],
                                      {'numerator': SUM('numerator-term'),
                                       'denominator': SUM('denominator-term')})
        # provider1 provider2 numerator denominator                                                                                    

        similarity['similarity'] = similarity.apply(lambda row:
                          divide(row['numerator'], row['denominator']))
        # provider1 provider2 numerator denominator similarity                                                                         

        mean_similarity = similarity['similarity'].mean()
        similarity['reg-similarity'] = similarity.apply(lambda row:
                         regularize(row['numerator'], row['denominator'], mean_similarity, k))

        del similarity['numerator']
        del similarity['denominator']
        # provider1 provider2 similarity reg-similarity                                                                                

        self.similarity = similarity
        return similarity

    # scores should look like                                                                                                          
    # <provider-col> <score-col> 
    # It is a way of assigining weights to providers as a whole.                                                                       
    # The weighting above is a way of assigning weights to individual                                                                  
    # provider-patient relationships                                                                                                   
    def make_provider_score_weighted(self, scores, provider_col, score_col):
        """ Weight the similarity by the provider score. """
        weighted_similarity = self.similarity.join(scores, on={'provider2': provider_col})
        # provider1 provider2 similarity reg-similarity <score-col>                                                                    
        weighted_similarity['scaled-similarity'] = weighted_similarity.apply(
            lambda row: row['reg-similarity'] * row[score_col])
        # provider1 provider2 similarity reg-similarity <score-col> scaled-similarity                                                  

        res = weighted_similarity.groupby('provider1', {'score': SUM('scaled-similarity')})
        res.rename({'provider1': 'provider'})
        # provider score                                                                                                               
        return res

    def make_provider_score_unweighted(self):
        """ Make the similarity only from the number of shared patients. """
        res = self.similarity.groupby('provider1', {'score': SUM('reg-similarity')})
        res.rename({'provider1': 'provider'})
        # provider score                                                                                                               
        return res
```

```python
# Compute M-transpose * M using scipy sparse matrix package
def mtm(similarity):
    """ Given a similarity matrix M in edge-list form, compute M-transpose * M in similar form.                                        
                                                                                                                                       
    Transform to scipy.sparse matrix to do the math."""

    # Step 1: transform the coordinates to integer                                                                                     
    # build a map                                                                                                                      
    providers = similarity['provider1']
    providers = providers.append(similarity['provider2'])
    providers = providers.unique()

    # number the rows                                                                                                                  
    provider_array = SFrame({'provider': providers})
    provider_array.add_row_number()

    provider_map = {}
    for row in provider_array:
        provider = row['provider']
        id = row['id']
        provider_map[provider] = id

    # build sparse matrix                                                                                                              
    size = len(providers)
    m = scipy.sparse.dok_matrix((size, size))
    for row in similarity:
        p1 = provider_map[row['provider1']]
        p2 = provider_map[row['provider2']]
        s = row['similarity']
        m[p1, p2] = s
        m[p2, p1] = s

    del provider_map

    mtm = m.transpose() * m

    # transform to coo to get the result                                                                                               
    mtm_coo = mtm.asformat('coo')
    res = SFrame({'provider1_index': mtm_coo.row, 'provider2_index': mtm_coo.col, 'similarity': mtm_coo.data})

    # clean up in case memory runs low                                                                                                 
    del mtm
    del mtm_coo

    # create reverse map                                                                                                               
    provider_reverse_map = {}
    for row in provider_array:
        provider = row['provider']
        id = row['id']
        provider_reverse_map[id] = provider

    res['provider1'] = res['provider1_index'].apply(lambda x: provider_reverse_map[x])
    res['provider2'] = res['provider2_index'].apply(lambda x: provider_reverse_map[x])
    del res['provider1_index']
    del res['provider2_index']
    return res

```
